# **Hyperparameter Tuning**

```python
# Step 1: Import required libraries
# from sklearn.datasets import load_diabetes
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.metrics import r2_score



# Step 3: Define models
models = {
    'KNN Regressor': KNeighborsRegressor(),
    'Linear Regression': LinearRegression(),
    'Decision Tree Regressor': DecisionTreeRegressor(random_state=42),
    'Random Forest Regressor': RandomForestRegressor(random_state=42),
    'AdaBoost Regressor': AdaBoostRegressor(random_state=42)
}

# Step 4: Define parameter grids for each model
param_grids = {
    'KNN Regressor': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'p': [1, 2]
    },
    'Linear Regression': {
        'fit_intercept': [True, False],
        'copy_X': [True, False]
    },
    'Decision Tree Regressor': {
        'criterion': ['squared_error', 'friedman_mse'],
        'max_depth': [None, 5, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'Random Forest Regressor': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'bootstrap': [True, False]
    },
    'AdaBoost Regressor': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 1.0]
    }
}

# Step 5: Perform Grid Search for each model
best_models = {}

for name, model in models.items():
    print(f"\n🔍 Tuning {name}...")

    grid = GridSearchCV(
        estimator=model,
        param_grid=param_grids[name],
        cv=5,
        scoring='r2',
        n_jobs=-1
    )

    grid.fit(x_train,y_train)
    best_models[name] = grid.best_estimator_

    print(f"✅ Best Parameters for {name}: {grid.best_params_}")
    print(f"🏆 Best R² Score (CV): {grid.best_score_:.4f}")

# Step 6: Evaluate the best models on test data
print("\n📊 Final Model Performance on Test Set:")
for name, model in best_models.items():
    y_pred = model.predict(x_test)
    print(f"{name}: R² Score = {r2_score(y_test, y_pred):.4f}")
```

## **DECISION TREE ALGORITHM**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python


# Create the model with key hyperparameters
deci = DecisionTreeRegressor(
    criterion='squared_error',   # or 'friedman_mse' / 'absolute_error'
    splitter='best',             # 'best' or 'random'
    max_depth=5,                 # limits tree depth (controls overfitting)
    min_samples_split=4,         # minimum samples to split a node
    min_samples_leaf=2,          # minimum samples in each leaf node
    max_features=None,           # number of features to consider for best split
    random_state=42,             # ensures reproducibility
    max_leaf_nodes=None,         # limit number of leaf nodes (optional)
    ccp_alpha=0.0                # pruning parameter (larger → simpler tree)
)
```
